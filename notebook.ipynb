# 2. Conceptos y algoritmos

## 2.1. Método de Clusterización k-means



La técnica de clusterización no supervisada k-means es un algoritmo popular para agrupar datos en clusters basados en la similitud. Se considera un método de aprendizaje automático no supervisado, ya que no requiere etiquetas de clase predefinidas para los datos. El algoritmo k-means fue desarrollado por [Stuart Lloyd](https://users.cs.utah.edu/~jeffp/teaching/cs5140-S14/cs5140/L10-kmeans.pdf), en 1957.

1. Funcionamiento del algoritmo

   - Selección de k: Se elige un número k de clusters que se desea obtener.
   - Inicialización de centroides: Se seleccionan al azar k puntos de datos como centroides iniciales de los clusters.
   - Asignación de clusters: Se asigna cada punto de datos al cluster con el centroide más cercano.
   - Actualización de centroides: Se recalculan los centroides como la media de los puntos de datos asignados a cada cluster.
   - Repetición de pasos 3 y 4: Se repiten los pasos 3 y 4 hasta que los centroides no cambien significativamente o se alcance un número máximo de iteraciones.

2. Ventajas de k-means

   - Es un algoritmo simple y eficiente.
   - Es fácil de implementar y entender.
   - Es escalable a grandes conjuntos de datos.

3. Desventajas de k-means

   - La elección del valor de k puede ser difícil.
   - El algoritmo puede ser sensible a valores atípicos en los datos.
   - El algoritmo puede converger a soluciones subóptimas.


## 2.2. Método de Clusterización Jerarquico

La técnica de [clusterización jerárquica](https://dl.acm.org/doi/abs/10.1145/331499.331504) es un método para agrupar datos en una jerarquía de clusters. Este método comienza con cada punto como un cluster individual y luego combina clusters de forma iterativa hasta que se forma un único cluster que contiene todos los puntos. La clusterización jerárquica se puede dividir en dos tipos principales: aglomerativa y divisiva.

1. Clusterización jerárquica aglomerativa

* Algoritmo:

    - Se calcula la distancia entre todos los pares de puntos.
    - Se fusionan los dos puntos más cercanos en un solo cluster.
    - Se actualiza la matriz de distancias para reflejar la fusión.
    - Se repiten los pasos 2 y 3 hasta que solo quede un cluster.
    - Se crea un dendrograma para visualizar la jerarquía de clusters.

* Ventajas:
    - Es simple y eficiente.
    - Es capaz de encontrar clusters de forma irregular.
    - Es robusta a la presencia de ruido en los datos.

* Desventajas:
    - La elección de la medida de distancia puede ser difícil.
    - El algoritmo puede ser lento para conjuntos de datos grandes.

2. Clusterización jerárquica divisiva

* Algoritmo:
   - Se comienza con un único cluster que contiene todos los puntos.
   - Se divide el cluster en dos subclusters de forma iterativa.
   - Se repiten los pasos 2 y 3 hasta que cada punto sea un cluster individual.
   - Se crea un dendrograma para visualizar la jerarquía de clusters.

* Ventajas:
   - Es flexible y permite diferentes criterios de división.
   - Es capaz de encontrar clusters de forma irregular.

* Desventajas:
   - Puede ser sensible a la elección del criterio de división.
   - Puede ser computacionalmente costoso para conjuntos de datos grandes.


## 2.3. Algoritmo de Kruskal

El algoritmo de Kruskal es un algoritmo voraz que se utiliza para encontrar un árbol recubridor mínimo (ARM) en un grafo conexo y ponderado. Un ARM es un subconjunto de aristas que conecta todos los vértices del grafo sin formar ciclos y con la suma de pesos de las aristas mínima. Este algoritmo fue presentado por Joseph Kruskal en su artículo ["On the shortest spanning subtree of a graph"](http://www.cs.yale.edu/homes/lans/readings/routing/kruskal-1956.pdf), publicado en 1956 .

1. Funcionamiento del algoritmo

   - Ordenar las aristas: Se ordenan todas las aristas del grafo por su peso de menor a mayor.
   - Recorrido de las aristas: Se recorren las aristas ordenadas una a una.
   - Comprobar ciclo: Para cada arista, se comprueba si al añadirla al árbol actual se crea un ciclo. Si se crea un ciclo, la arista se descarta.
   - Añadir arista: Si no se crea un ciclo, la arista se añade al árbol actual.
   - Repetir hasta completar el árbol: Se repiten los pasos 3 y 4 hasta que el árbol actual tenga todos los vértices del grafo conectados.

2. Eficiencia del algoritmo

El algoritmo de Kruskal tiene una complejidad temporal de O(E log E), donde E es el número de aristas del grafo. Esta complejidad se debe principalmente al ordenamiento de las aristas. En la práctica, el algoritmo de Kruskal es uno de los algoritmos más eficientes para encontrar un ARM.

3. Aplicaciones del algoritmo

   - Redes de comunicación: Se puede utilizar para encontrar la ruta de menor costo para conectar todos los nodos de una red.
   - Diseño de circuitos: Se puede utilizar para encontrar la ruta de menor costo para conectar todos los componentes de un circuito.
   - Planificación de rutas: Se puede utilizar para encontrar la ruta de menor distancia para visitar un conjunto de ciudades.

### 5.1.1. Implementación k-means

````python
# Se seleccionan las variables de estudio
X = df.loc[:, ['Lon', 'Lat']]

# Se aplica algoritmo k-means con k = 100
kmeans = cluster.KMeans(n_clusters=100, max_iter=1000, random_state=seed).fit(X)
````

A continuación vamos a implementar una función para generar un gráfico de dispersión que permita visualizar los cluster genrados

````python
def cluster_graph(X_data: pd.Series, labels: pd.Series, 
                  title_graph: str, map_color: str='tab20'):
    """
    Genera gráfico de dispersión del set de datos con las etiquetas
    de cada cluster 
    :param X_data: pd.Series
            Datos con las variables de estudio.
    :param labels: pd.Series
             Etiquetas obtenidas por el método de clúster.            
    :param title_graph: str
            Título del gráfico.
    :param map_color: str
            mapa de color
            default: 'tab20'
    :return: None
    """
    plt.title(title_graph)
    plt.scatter(X_data.iloc[:,0], X_data.iloc[:,1], s=15, c=labels, alpha=.5, cmap=map_color)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

    return
````

````python
# Se grafica el set de datos con las etiquetas generadas por el algoritmo k-means
cluster_graph(X, kmeans.labels_, 'Groups k-means')
````

En el gráfico anterior observamos  que los clusters estan claramente definidos y delimitados.

### 5.1.2. **Implementación Algoritmos jerárquicos**

En esta sección vamos aplicar los algoritmos de clusterización jerárquicos, teniendo en cuenta diferentes tipos de linkage. Para ello implementaremos una funciíon que permita aplicar los difererentes métodos al conjunto de datos.

````python
def cluster_hierarchy(X_data: pd.Series) -> None:
    """
    Genera dendograma para cada criterio de enlace
    :param X_data: pd.Series
            Series de datos con las variables de estudio.
    :return: None
    """
    method_linkage = ['single', 'complete', 'centroid', 'median', 'average', 'ward']
    plt.figure(6, figsize=(16, 8))
    plt.clf()
    for k, v in enumerate(method_linkage):
        plt.subplot(2, 3, k+1)
        _ = sch.dendrogram(sch.linkage(X_data, method = v))
        plt.gca().axes.get_xaxis().set_visible(False)
        plt.title(v)
    plt.show()
````

````python
# Se generan dendogramas para cada criterio de enlace
cluster_hierarchy(X)
````

A partir del eje vertical en los dendogramas, definimos los rangos de corte aproximados para cada método de modo que se puedan separar 100 cluster.

````python
range_cluster = {'single': [0.02, 0.00001],
                 'complete': [0.02, 0.0001],
                 'centroid': [0.01, 0.0001],
                 'median': [0.01, 0.0001],
                 'average': [0.01, 0.0001],
                 'ward': [0.04, 0.0001]}
````

A continuación se itera sobre este rango, para cada criterio hasta que obtengamos 100 clusters

````python
# Se obtienen los puntos de corte del eje vertical del dendograma,  para cada criterio de enlace
# Considerando que el valor máximo número de clusters es 100.
method_linkage = ['single', 'complete', 'centroid', 'median', 'average', 'ward']
dict_labels = {}
for k, v in enumerate(method_linkage):
    cluster_sch = sch.linkage(X, method=v)
    for i in reversed(np.arange(0.0, range_cluster[v][0], range_cluster[v][1])):
        labels_shc_c = sch.fcluster(Z=cluster_sch, t=i, criterion='distance')
        if len(np.unique(labels_shc_c)) >= 100:
            dict_labels.update({v: [len(np.unique(labels_shc_c)), i]})
            break
            
dict_labels
````

Con la información anterior y con la función fcluster, extraemos las etiquetas de los clusters para cada criterio de enlace y graficamos los resultados.

````python
# Se grafica el set de datos con las etiquetas generadas para cada criterio de enlace
plt.figure(6, figsize=(16, 8))
plt.clf()
for k, v in enumerate(method_linkage):
    plt.subplot(2, 3, k+1)
    cluster_sch = sch.linkage(X, method=v)
    labels_shc_c = sch.fcluster(Z=cluster_sch, t=dict_labels[v][1], criterion='distance')
    plt.scatter(X.iloc[:,0], X.iloc[:,1], s=15, c=labels_shc_c, alpha=.5, cmap='tab20')
    plt.title(v)
plt.show()
````

En los gráficos anteriores observamos que el método jerárquico, con el criterio de enlace single, no evidencia un agrupamiento definido. Los demás criterios de enlace, aunque se observa separación entre cluster, no son uniformes y en algunos casos un cluster contiene a otro (centroid y median).

De acuerdo a los resultados anteriores el método de clustering con el que obtenemos agrupaciones mas uniformes y  definidas es k-means. Dado que el problema radica en la creación de una red de transporte entre distintos puntos, conviene que los centroides (puntos de referencia para la creación de las rutas) generados por el método de clusterización, deben pertenecer a regiones/grupos claramente separados y uniformes.

## 5.2. Análisis de grupos y preparación de datos

Antes de continuar con el diseño de la red de transporte hay que validar que los grupos tienen sentido.

Lo primero es estimar la densidad de cada grupo encontrado, a través del cálculo de area que agrupa cada cluster.
Una aproximación es la siguiente:

- Calcular el area aproximada a través de (lat_max - lat_min) * (lon_max - lon_min), teniendo cuidado con los valores negativos, de los puntos que pertenecen al grupo.

- Calcular el número de puntos que pertenecen a cada grupo.

- Calcular el ratio número de puntos por área cubierta de cada cluster.

- Finalmente visualizar para cada cluster el resultado del número de puntos que pertenecen a cada grupo en un diagrama de barras.

````python
# Se extraen las etiquetas de los cluster generadas por el algoritmo k-means
df['labels'] = kmeans.labels_
list_cluster = list(df.groupby(['labels']))
````

````python
def info_cluster(data: pd.DataFrame) -> list:
    """
    Crea lista con información asociada a cluster:
    area, número de puntos y el rattio area/# puntos, para cada cluster
    :param data: pd.DataFrame
            Dataframe con las variables de estudio.
    :return: list
                Contiene el area, número de puntos y el rattio area/# puntos
                para cada cluster 
    """
    lat_dist = data['Lat'].max() - data['Lat'].min()
    lon_dist = abs(data['Lon'].max() - data['Lon'].min())

    area_cluster = lat_dist*lon_dist
    num_points = len(data['Lat'])
    rt_point_area = num_points/area_cluster

    return [area_cluster, num_points, rt_point_area]
````

````python
# Extrae información de cada cluster y lo almacena en un dicionario
dict_cluster_info = {k: info_cluster(v[1]) for k, v in enumerate(list_cluster)}
````

````python
# Transforma el diccionario en dataframe
df_cluster_info = pd.DataFrame(dict_cluster_info).T
df_cluster_info.columns = ['area_cluster', 'num_points', 'rt_point_area']
df_cluster_info.head(5)
````

````python
# Genera gráfico que visualiza el ratio area/#puntos para cada cluster
_ = df_cluster_info['rt_point_area'].plot(kind='bar', figsize = (18, 12))
plt.xlabel('Número de clusters')
plt.ylabel('Ratio Área/Número de puntos por cluster')
````

En el gráfico apreciamos que algunos clúster tienen una baja proporción entre el número de puntos y el área,
por lo tanto, los valores etiquetados que pertencen a dicho cluster se eliminarán del conjunto de datos. En este sentido, descartaremos lo cluster cuya proporción numero de puntos sobre area, sea menor el cuantil del 5%.

````python
df_row_drop = df_cluster_info[df_cluster_info['rt_point_area'] < df_cluster_info['rt_point_area'].quantile(0.05)]
df = df.drop(index=list(df_row_drop.index))
````

## 5.3. Aplicación del algoritmo Kruskal para crear la red de transporte

A continuación procedemos ha implementar una solución que minimice la extensión de la red tratando de dar cobertura al mayor número de peticiones posibles.

Para ello se pueden usar distintas estrategias. Entre ellas la familia *MST* o [*Minimum Spanning Tree*](https://en.wikipedia.org/wiki/Minimum_spanning_tree), con algoritmos como el de [*Kruskal*](https://en.wikipedia.org/wiki/Kruskal%27s_algorithm) o *Prim*. El algoritmo de *Kruskal* tiene la ventaja de ser muy rápido por ser *greedy* o voraz (subóptimo) y, además, evita ciclos. Un beneficio extra, ya que en nuestro caso establecería enlaces redundantes, los cuales queremos evitar por el elevado coste.

Los pasos son los siguientes:
1. Preparar los datos para el grafo.
2. Aplicar el algoritmo de Kruskal.
3. Analizar los resultados y decidir el diseño final de la red.

En primer lugar vamos a crear un dataframe en el que haya una fila por cluster y ésta contenga: <b>id</b> del cluster (numérico), <b>latitud</b> y <b>longitud</b> (representativa del cluster), y <b>número de puntos</b> pertenecientes al cluster.

````python
# Se genera dataframe con los centroidos y el número de puntos de cada cluster
data_grafo = pd.concat([pd.DataFrame(kmeans.cluster_centers_, columns=['Lon', 'Lat']), df_cluster_info.loc[:, 'num_points']], axis=1)
data_grafo.reset_index(inplace=True)
data_grafo.rename(columns={'index': 'id'}, inplace=True)
data_grafo.head(5)
````

Para crear el grafo no dirigido (si 1 está conectado con 2 se asume la dirección inversa) a partir del cual el algoritmo de *Kruskal* seleccionará los enlaces es necesario indicarle las posibles aristas del grafo.

Para crear el grafo simplemente hay que instanciar la clase `Graph` del módulo *kruskal* que hemos modificado a partir de [aquí](https://www.pythonpool.com/kruskals-algorithm-python/). Indicando en el parámetro `vertex` de su constructor el id más alto de cluster +1. Posteriormente se pueden añadir aristas al grafo indicando primero los dos ids de clusters que une el tramo y por último su peso que viene indicado por la definición de *weight* más adelante.

Por último, se calcula el resultado de *Kruskal* sobre el grafo haciendo una llamada al método `kruskal`, que también dispone de un parámetro `max_links` si sólo deseamos que se generen tantos tramos como se indiquen (útil más adelante).

Un ejemplo concreto sería el siguiente:

```
g = kruskal.Graph(vertex=5)
g.add_edge(0, 1, 8)
...
g.add_edge(3, 4, 7)
g.kruskal()
```

Debido al alto coste es necesario justificar la construcción de cada tramo del *hypermegaminiloop*, por lo que vamos a emplear una métrica que pese (*weight*) cada tramo (unión entre 2 clusters cualquiera $i$ y $j$):

$$
weight(i, j) = -\frac{n_i+n_j}{dist(i, j)}
$$

Donde:
 - $n_i$ es el número de puntos del cluster $i$, de la misma manera $n_j$ es el número de puntos del cluster $j$.
 - $dist(i, j)$ es la distancia geográfica en km entre el centro del cluster $i$ y el cluster $j$.
 
La distancia geográfica entre 2 puntos cualesquiera no es la distancia euclídea, puesto que la tierra tiene curvatura. Para ello puedes usar el módulo [haversine](https://pypi.org/project/haversine/) (ya importado como *hs* al inicio de la PEC).

Por último, resaltar que el peso final es negativo porque el algoritmo de *Kruskal* ve este peso como un coste que trata de minimizar. Por nuestra parte, deseamos incentivar que a medida que más alto sea el ratio descrito más atractivo le sea a *Kruskal*, por eso se hace negativo.

A continuación, vamos a crear un grafo siguiendo el código de ejemplo en el que aparezcan todos los pares de clusters posibles (no hace falta repetir los pares porque se trata de un grafo no dirigido donde A-B también equivale a B-A), donde el peso (último parámetro del método add_edge) se calcule tal y como se describe en la fórmula de weight.

````python
g = kruskal.Graph(vertex=data_grafo.id.iloc[-1]+1)
````

````python
# Se crea los pares del grafo no dirido
lenght = data_grafo.id.iloc[-1]+1

dict_uv = {}
count = 0
for i in range(0, lenght):
    for j in range(i+1, lenght):
        dict_uv.update({count: [i, j]})
        count = count + 1
````

````python
# Se crea diccionario que contiene la distancia heversine para cada par de clusters
dict_weigth = {}
for i in dict_uv:
    cluster_points = data_grafo.iloc[dict_uv[i][0], [3]] + data_grafo.iloc[dict_uv[i][1], [3]]
    cluster_dist = haversine(data_grafo.iloc[dict_uv[i][0], [1, 2]], data_grafo.iloc[dict_uv[i][1], [1, 2]])
    dict_weigth.update({i: -cluster_points/cluster_dist}) 
````

````python
# Se crea el grafo utilizando cada par de cluster y las distancias heversine asociadas
for i in dict_uv:
    g.add_edge(dict_uv[i][0], dict_uv[i][1], dict_weigth[i][0])
````

Aplicamos el algoritmo de Kruskal llamando al método kruskal del grafo variando el número de ejes en función de tu criterio (compromiso entre extensión y coste).
